```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('dataset.csv')
```

```python
movies.info()
```

```python
movies.head()
```

```python
movies['tags'] = movies['genre'] + movies['overview']
```

```python
movies.head()
```

```python
new_df = movies[['id','title','genre', 'overview', 'tags']]
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')). toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
new_df[new_df['title'] == 'The Shawshank Redemption']
```

```python
dist = sorted(list(enumerate(sim[2])), reverse=True, key=lambda vector: vector[1])
for i in dist[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
def recommend(movies):
  index = new_df[new_df['title'] == movies].index[0]
  distance = sorted(list(enumerate(sim[index])), reverse=True, key=lambda vector: vector[1])
  for i in distance[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("House of Gucci")
```
